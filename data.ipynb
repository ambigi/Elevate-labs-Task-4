```python
import pandas as pd
import random
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
```

```python

products = ["Laptop", "Phone", "Tablet", "Headphones", "Monitor"]
regions = ["North", "South", "East", "West"]

rows = []
start_date = datetime(2025, 9, 1)

for i in range(50):
    date = start_date + timedelta(days=i % 10)
    product = random.choice(products)
    region = random.choice(regions)
    sales = random.randint(500, 3000)
    rows.append([date.strftime("%Y-%m-%d"), product, region, sales])

df = pd.DataFrame(rows, columns=["Date", "Product", "Region", "Sales"])
df.to_csv("sales.csv", index=False)
```

```python

data = pd.read_csv("sales.csv")
print("\nPreview of Data:")
print(data.head())

print("\nTotal Sales by Region:")
print(data.groupby("Region")["Sales"].sum())

print("\nTotal Sales by Product:")
print(data.groupby("Product")["Sales"].sum())
```

```python

data.groupby("Region")["Sales"].sum().plot(kind="bar", title="Sales by Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
data.groupby("Product")["Sales"].sum().plot(kind="pie", title="Sales by Product", autopct='%1.1f%%')
plt.ylabel("")
plt.show()
```
